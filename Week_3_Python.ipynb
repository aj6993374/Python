# Label  based indexing (.loc[])

- .loc[] is used for selecting rows and columns by labels 
- we can use single lables, list of labels or slices of labels

```python
# Selecting rows based on Label
students=pd.read_csv('data/student.csv')
first_rows=students.loc[1]
first_two_rows=students.loc[0:2]
print(first_rows)
print(first_two_rows)

# # seledting specific rows to columns 
first_two_stds_with_name_to_marks =students.loc[0:2,'name':'mark']
print(first_two_stds_with_name_to_marks) # rows 0 to 2, columns name to marks 

# select only particular rows and columns
first_and_second_rows_with_name_and_gender=students.loc[[5, 20],['name','gender']]

#Row 5 and columns name and gender 
first_and_second_rows_with_name_and_gender


```

# integer-location Based Indexing 
- .iloc[] is used for selecting rows and columns by integer index positions
- similar to .iloc[], it is used single integer, list of integers, or slices 

```python
# selecting rows based on index positions 
print(students.iloc[1])
print(students.iloc[0:3])

#selecting specific rows and columns 
print (students.iloc[0:2, 0:2]) #rows 0 to 1, columns 0 to 2
print (students.iloc[[0,2],[0,2]]) #rows 0 and 2, columns 0 and 2
```

# Single Value Access  (.at[], and .iat[])
- .at[] is used for accessing a single scalar valued by label.
- .iat is used for accessign for index position. 


```python
# using .at[] to access single value by label
print(students.at[1,'name'])  #value in row 1 and column name 

# Using .iat[] to access single value by position
print(students.iat[1, 1])                 # Value in row 1, column 1
```
